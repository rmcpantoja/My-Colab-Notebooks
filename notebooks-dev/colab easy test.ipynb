## cómo desconectarse de la máquina:
1. Presiona la combinación de teclas control+inicio para ir al comienzo de toda la página. Navega hasta que te encuentres con la barra de menú.
2. Deberás seleccionar el elemento entorno de ejecución con enter.
3. Selecciona "administrar secciones". Si usas NVDA, apunta el elemento seleccionado al mouse usando la combinación Shift+NVDA+M en distribuciones portátiles, y a continuación, clic izquierdo.
4. Aparecerá un cuadro de diálogo con las sesiones. Solamente tendremos la sesión del cuaderno de la voz que estamos usando en este momento (a menos que esté ejecutando otros notebooks) Deberás finalizar la sesión de este cuaderno presionando el botón respectivo.
5. Si nos pregunta si confirmamos que deseamos finalizar la sesión, presionamos en finalizar nuevamente y nuestro entorno se desconectará.
